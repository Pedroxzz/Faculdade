```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Funo para carregar uma lista de imagens coloridas
def load_color_images(image_paths):
    images = [cv2.imread(path, cv2.IMREAD_COLOR) for path in image_paths]
    return images

# Funo para calcular a mdia de vrias imagens coloridas
def average_color_images(images):
    # Converter a lista de imagens para um array numpy e calcular a mdia por canal (RGB)
    average_image = np.mean(images, axis=0).astype(np.uint8)  # Corrigido para np.uint8
    return average_image

# Lista de caminhos das 5 imagens ruidosas coloridas
image_paths = [
    'passaros_R1.png',
    'passaros_R2.png',
    'passaros_R3.png',
    'passaros_R4.png',
    'passaros_R5.png'
]

# Carregar as imagens coloridas
images = load_color_images(image_paths)

# Calcular a imagem mdia (reduo de ruido)
average_image = average_color_images(images)

# Converter a imagem de BGR (padro OpenCV) para RGB (padro matplotlib)
average_image_rgb = cv2.cvtColor(average_image, cv2.COLOR_BGR2RGB)  # Corrigido para cv2.COLOR_BGR2RGB

# Exibir a imagem resultante
plt.figure(figsize=(6,6))
plt.title("Imagem Mdia Colorida (Reduo de Rudo)")
plt.imshow(average_image_rgb)
plt.axis('off')  # Ocultar os eixos
plt.show()

# Salvar a imagem resultante
cv2.imwrite('average_color_image_result.jpg', average_image)
```
