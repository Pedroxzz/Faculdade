# EXERCÍCIOS


1.   Crie uma agenda de contatos contendo nome e celular de **n** pessoas, armazenados em um dicionário. Em seguida crie métodos para consultar, inserir, alterar e deletar itens neste dicionário. Os dados devem ser lidos pelo teclado.

2.   Faça um programa que carrege dados em um dicionário, onde a chave deverá ser um número aleatório entre 0 e 100 e o valor deverá ser uma palavra pré-cadastrada em uma tupla.

3.   Cadastre 5 funcionários em um dicionário, pelo teclado, com as seguintes chaves: nome, idade, cargo, salário. No final, mostre o conteúdo do dicionário e a média dos salários.

4. Carregue um vetor com 5 números inteiros e mostre o maior deles, sem usar a função **max()**.

5. Carregue um vetor com 3 notas e mostre a média das notas 

6. Cadastre em vetores os nomes e as idades de 5 pessoas. Em seguida, mostre a idade e a pessoa mais velha cadastrada.

```python
#1

agenda = {}

def inserir_contato():
    nome = input("Digite o nome do contato: ")
    celular = input("Digite o celular do contato: ")
    agenda[nome] = celular
    print("Contato adicionado.")

def consultar_contato():
    nome = input("Digite o nome do contato:")
    if nome in agenda:
        print(f"Nome: {nome} - Celular: {agenda[nome]}")
    else:
        print("Contato não encontrado")

def alterar_contato():
    nome = input("Digite o nome do contato que deseja alterar: ")
    if nome in agenda:
        celular = input("Digite o novo celular do contato: ")
        agenda[nome] = celular
        print("Contato alterado com sucesso!")
    else:
        print("Contato não encontrado.")

def deletar_contato():
    nome = input("Digite o nome do contato que deseja deletar: ")
    if nome in agenda:
        del agenda[nome]
        print("Contato deletado com sucesso!")
    else:
        print("Contato não encontrado.")

while True:
    print("=== AGENDA DE CONTATOS ===")
    print("1 - Inserir novo contato")
    print("2 - Consultar contato")
    print("3 - Alterar contato")
    print("4 - Deletar contato")
    print("5 - Sair")
    opcao = input("Digite a opção desejada: ")
    
    if opcao == "1":
        inserir_contato()
    elif opcao == "2":
        consultar_contato()
    elif opcao == "3":
        alterar_contato()
    elif opcao == "4":
        deletar_contato()
    elif opcao == "5":
        break
    else:
        print("Opção inválida. Tente novamente.")
```

```python
#2

import random

palavras = ("café","leite", "pessoa", "arvore", "carro", "radar", "maça", "banana", "folha", "faixa", "policial", "trem", "onibus")
dicionario = {}

for i in range(10):
    chave = random.randint(0,100)
    palavra = random.choice(palavras)
    dicionario[chave] = palavra
print(dicionario)
    
```

```python
#3

funcionarios = {}

for i in range(1):
    print(f"Cadastro do {i+1}º funcionário")
    nome = input("Nome: ")
    idade = int(input("Idade: "))
    cargo = input("Cargo: ")
    salario = float(input("Salário: "))

    funcionarios[i+1] = {'nome':nome, 'idade':idade, 'cargo':cargo, 'salario':salario}

media_salarios = sum(funcionarios[i]['salario'] for i in range(1, 6)) / 5
print(f"Média dos salários: R$ {media_salarios:.2f}")
```

```python
#4

vetor = []
maior = 0
for i in range(5):
    numero = int(input(f"Digite o {i+1}º número: "))
    vetor.append(numero)
    if numero > maior:
        maior = numero

print(vetor)
print(maior)
```

```python
#5
notas = []
media = 0

for i in range(3):
    nota = float(input(f'Digite a {i+1}ª nota: '))
    media += nota
    notas.append(nota)
media = media /3
print(f'A média das notas é: {media:.2f}')
```

```python
#6 

pessoas = []

for i in range(5):
    nome = input("Digite o nome da pessoa: ")
    idade = int(input("Digite a idade da pessoa: "))

    pessoas.append({'nome': nome, 'idade': idade})


mais_velha = pessoas[0]
for pessoa in pessoas:
    if pessoa['idade'] > mais_velha['idade']:
        mais_velha = pessoa


print(f"A pessoa mais velha é {mais_velha['nome']}, com {mais_velha['idade']} anos.")
```
